### search entry and results

```python
driver.get('https://7esl.com/texting-abbreviations/')
```

```python
# create empty lists to hold data
abb = []
exp = []
```

```python
abb = []
exp = []
ini = []
a = driver.find_element_by_css_selector("div.entry-content")
table = a.find_elements_by_tag_name("table")
for t in table:
    body = t.find_element_by_tag_name('tbody')
    rows=body.find_elements_by_tag_name('tr')
    rows = rows[1:]
    for x in rows:
        text = x.get_attribute('innerText').split('\t')
        abb.append(text)
        ini.append(text[0])
        exp.append(text[1])
        

```

```python
idi = pd.DataFrame()
idi['a'] = ini
idi['b'] = exp
idi.head()
```

```python
mask = (idi['a'].str.len()>=2) 
df = idi.loc[mask]
print(df)
```
